### Multilingual multi-backend compilation example

The QLLVM compiler supports multiple quantum programming languages (Qiskit, Criq, Qpanda, openQASM) and various quantum backends (qasm simulator, OriginIR of Qiskit, qcis of Tianyan). Here is a compilation example.

### 1. The input is a program written in Qiskit programming language.

#### (1)  The output is a QASM program.

```python
from qiskit import QuantumCircuit
from qllvm import compile

qc = QuantumCircuit(3, 3)

qc.h(0)
qc.cx(0, 1) 
qc.cx(0, 2)      

qc.measure([0,1,2], [0,1,2])
compile(qc,input='qiskit',out_backend='qasm-backend',basicgate='[rx,ry,cx]',optlevel='1',output_path = './test')

print("=====================Transpiled QASM program=====================")
with open('./test.qasm','r') as f:
    print(f.read())
```

#### (2)  The output is a QCIS program.

```python
from qiskit import QuantumCircuit
from qllvm import compile

qc = QuantumCircuit(3, 3)

qc.h(0)
qc.cx(0, 1) 
qc.cx(0, 2)      

qc.measure([0,1,2], [0,1,2])

compile(qc,input='qiskit',out_backend='tianyan',optlevel='1',output_path = './test')

print("=====================Transpiled QCIS program=====================")
with open('./test.qcis','r') as f:
    print(f.read())
```

#### (3)  The output is a OriginIR program.

```python
from qiskit import QuantumCircuit
from qllvm import compile

qc = QuantumCircuit(3, 3)

qc.h(0)
qc.cx(0, 1) 
qc.cx(0, 2)      

qc.measure([0,1,2], [0,1,2])

compile(qc,input='qiskit',out_backend='originquantum',optlevel='1',output_path = './test')

print("=====================Transpiled OriginIR program=====================")
with open('./test.py','r') as f:
    print(f.read())
```

### 2. The input is a program written in Qpanda programming language.

#### (1)  The output is a QASM program.

```python
from pyqpanda3.core import QProg,measure,draw_qprog,CPUQVM,H,CNOT
from qllvm import compile

prog = QProg()
prog << H(0)
for i in range(1, 5):
    prog << CNOT(i - 1, i)
for i in range(5):
    prog << measure(i, i)

compile(prog,input='qprog',out_backend='qasm-backend',basicgate='[rx,ry,cx]',optlevel='1',output_path = './test')
print("=====================Transpiled QASM program=====================")
with open('./test.qasm','r') as f:
    print(f.read())
```

### 3. The input is a program written in QASM programming language.

#### (1)  The output is a QASM program.

```python
from qllvm import compile

qasm_str = '''
OPENQASM 2.0;
include "qelib1.inc";
qreg q[3];
creg meas[3];
h q[2];
cx q[2],q[1];
cx q[1],q[0];
barrier q[0],q[1],q[2];
measure q[0] -> meas[0];
measure q[1] -> meas[1];
measure q[2] -> meas[2];
'''

compile(qasm_str,input='qasm',out_backend='qasm-backend',basicgate='[rx,ry,cx]',optlevel='1',output_path = './test')
print("=====================Transpiled QASM program=====================")
with open('./test.qasm','r') as f:
    print(f.read())
```

### 4. The input is a program written in Cirq programming language.

#### (1)  The output is a QASM program.

```python
import cirq
from qllvm import compile

q0, q1 = cirq.LineQubit.range(2)

bell_circuit = cirq.Circuit()
bell_circuit.append(cirq.H(q0))
bell_circuit.append(cirq.CNOT(q0, q1))

bell_circuit.append(cirq.measure(q0, q1, key='result'))

compile(bell_circuit,input='cirq',out_backend='qasm-backend',basicgate='[rx,ry,cx]',optlevel='1',output_path = './test')
print("=====================Transpiled QASM program=====================")
with open('./test.qasm','r') as f:
    print(f.read())
```
